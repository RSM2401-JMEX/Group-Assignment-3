## Task 2a

```python
#i.Create a line chart showing the total annual import and export of all countries each year from 2010 to 2018
export = Trade.Flow =='Exports'
list_export = Trade[export]
plot_exp = list_export.groupby('TIME')['Value'].sum()

timport = Trade.Flow =='Imports'
list_import = Trade[timport]
plot_imp = list_import.groupby('TIME')['Value'].sum()

plt.plot(plot_exp)
plt.plot(plot_imp)
```

```python
#ii. Create a line chart showing the net export of all countries each year from 2010 to 2018
netexp = Trade.groupby(['TIME','Flow'])['Flow','Value'].sum()
netexp = netexp.unstack() #unstacking Flow
netexp['Net_Exports']=netexp.Value['Exports']-netexp.Value['Imports']
netexp.head()
```

```python
plot_netexp = netexp.Net_Exports
plt.plot(plot_netexp)
```

```python
#iii. Create a line chart showing the total GDP of all countries each year from 2010 to 2018
GDP_countries = GDP.groupby('TIME')['Value'].sum()
plt.plot(GDP_countries)
```

# Task 3

```python
#i. Create a stacked area chart, a stacked bar chart, and a normalized stacked bar chart showing 
#the total annual export for each country each year from 2010 to 2018.

#filtering exports
Trade_exp = Trade.Flow == 'Exports'
Trade2 = Trade[Trade_exp]

exp_country = Trade2.groupby(['TIME','LOCATION'])['Value'].sum()
exp_country = exp_country.unstack()
exp_country.head()
```

```python
#Area chart
exp_country.plot(kind='area', stacked=True, title="Annual Export for Each Country(Area)")
plt.legend(loc='lower left', ncol = 9, fontsize='xx-small')

#Bar chart
exp_country.plot(kind='bar', stacked=True, title="Annual Export for Each Country(Bar)")
plt.legend(loc='lower left', ncol = 9, fontsize='xx-small')

#Normalized Bar chart
exp_country.plot(kind='bar', stacked=True, title="Annual Export for Each Country(Normalized)")
plt.legend(loc='lower left', ncol = 9, fontsize='xx-small')

plt.tight_layout()
```

```python

#filtering imports
Trade_imp = Trade.Flow == 'Imports'
Trade3 = Trade[Trade_imp]

imp_country = Trade3.groupby(['TIME','LOCATION'])['Value'].sum()
imp_country = imp_country.unstack()
imp_country.head()
```

```python
#Area chart
imp_country.plot(kind='area', stacked=True, title="Annual Import for Each Country(Area)")

#Bar chart
imp_country.plot(kind='bar', stacked=True, title="Annual Import for Each Country(Bar)")

#Normalized Bar chart

```

# Task 4

```python
#Create an unstacked and a stacked bar chart showing the total annual net export for each country each year from 2010 to 2018

#Creating an unstacked data frame for net exports
netexp_country = Trade.groupby(['TIME','LOCATION','Flow'])['Flow','Value'].sum()
netexp_country = netexp_country.unstack() #unstacking Flow
netexp_country['Net_Exports']=netexp_country.Value['Exports']-netexp_country.Value['Imports']
netexp_country=netexp_country.loc[:,['TIME','LOCATION','Net_Exports']]
netexp_country = netexp_country.unstack() #unstacking Location
netexp_country.head()
```

```python
#Staked Bar Chart
netexp_country.plot(kind='bar', stacked=True, title="Annual Net Export for Each Country(Stacked)")
```

```python
#Unstaked Bar Chart
netexp_country.plot(kind='bar', stacked=False, title="Annual Net Export for Each Country(Unstacked)")
```
